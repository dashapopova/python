```python
import json
import urllib.request
import collections
import os
import re
from string import punctuation
import matplotlib.pyplot as plt

req = urllib.request.Request('https://api.vk.com/method/wall.get?owner_id=1&offset=0&count=10&v=5.92&access_token=8423c2448423c2448423c244d08441f2a1884238423c244dee1644d9e90529494134bf8') 
with urllib.request.urlopen(req) as response: 
    result = response.read().decode('utf-8')
data = json.loads(result)

i = 0
stroka = ""
while i < 10:
    stroka = stroka + data['response']['items'][i]['text']
    i = i + 1

massiv = stroka.strip().split()
massiv1 = [word.strip(punctuation) for word in massiv]
chast1 = collections.Counter(massiv1)

with open("C:\input.txt", "w", encoding="utf-8") as f:
    f.write(stroka)


os.system("C:\mystem.exe -lndc C:\input.txt C:\output.txt")


with open("C:\output.txt", "r", encoding="utf-8") as f:
    text = f.read()
    text2 = re.sub("\n", "", text)
    text3 = re.sub("\r", "", text2)
    text4 = text3.split("_")
chast2 = collections.Counter(text4)

freq = collections.Counter(text4).most_common(30)
print(freq)
```

```python
freq = dict(freq)
freq = {c : freq[c] for c in freq if freq[c] > 0 and len(c) > 3}

# сортируем словарь по значениям
sorted(freq.items(), key=lambda x: x[1], reverse=True)
```
